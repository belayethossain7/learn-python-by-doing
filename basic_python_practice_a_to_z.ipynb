Types of Inheritance :

Single inheritance--akta parent class ar akta child class thabe just.

Multiple Inheritance--- ak ar besi parent class theke 1 ta child class asbe

Multilevel Inheritance--- grandParent class theke parent class asbe and parent class theke child class asbe ,,ai level onujaiyee asbe

Hierarchical inheritance---akta parent class theke onk gula child class hobe

Hybrid Inheritance ---combine of all types of inheritance

```python
# single inheritance

class University:
  def name(self):
    print("University name is International University of Business Agriculture and Technology.")

class Cse(University):
  def d_name(self):
    print("Department of Computer Science and Engineering")
```

```python
d1=Cse()

d1.name()
d1.d_name()
```

```python
class Fruit:
  def name(self):
    print("Apple")

class Drinks:
  def nam(self):
    print("Mojo")

class Eat(Fruit,Drinks):
  def morning(self):
    print("Eating.....")

c= Eat()
```

```python
c.name()
c.nam()
c.morning()
```

```python
# Mutilevel inheritance

class Uni:
  def name1(self):
    print("IUBAT")

class Cse(Uni):
  def name2(self):
    print("Computer Science and Engineering")

class Stu(Cse):
  def name3(self):
    print("I am From IUBAT and Department is CSE")
```

```python
s=Stu()

s.name1()
s.name2()
s.name3()
```

```python
# hierarchical Inheritance

class DL:
  def intro(self):
    print("I am DEEP_Learing")

class CV(DL):
  def intro2(self):
    print("I am Computer Vision and part of Deep learing.")

class NLP(DL):
  def intro3(self):
    print("I am Naturl Language processign and part of Deep learing")
```

```python
cv1= CV()
cv1.intro()
cv1.intro2()
```

```python
nlp1=NLP()
nlp1.intro()
nlp1.intro3()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
language=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(language,people)
plt.xlabel('Language')
plt.ylabel('Number of People')
plt.title('Language used by visitor for Communication')
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
language=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
plt.title('Language use by visitor for communication')
ax.pie(people,labels=language,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2= plt.figure()
ax=fig2.add_axes([0,0,1,1])
plt.xlabel('angle')
plt.ylabel('sine and cosine value')
plt.title("Sine and Cosine wave")
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20* np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

```python
fig4=plt.figure()
ax=plt.axes(projection='3d')
z= 20* np.random.random(100)
x=np.sin(z)
y=np.cos(z)
plt.xlabel('x')
plt.ylabel('y')

plt.title('Example of 3d line plot')
ax.scatter(x,y,z,c=z,cmap='Reds')

plt.show()
```
